```python
r = 1
a = 0.3
x = np.linspace(-5*r,5*r,100)
softplus = lambda b: lambda x: (x-b < 10)*np.log1p(np.exp(x-b)) + (x-b >=10)*(x-b)
softplusm = lambda b: lambda x: softplus(b)(x)-softplus(b)(0)
plt.plot(x, softplusm(10)(x))
plt.plot(x, softplusm(4)(x))
plt.plot(x, softplusm(1)(x))
```

```python
swish = lambda b: lambda x:  x * expit(x-b)
plt.plot(x, swish(0.1)(2*x))
plt.plot(x, swish(1)(x))
plt.plot(x, swish(4)(x))
```
