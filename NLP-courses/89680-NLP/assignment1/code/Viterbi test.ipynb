```python
# now create one big state matrix
def create_big_state_matrix(layer_distance):
    ls = 4
    layers = 8
    cities_west_start_index=0
    cities_east_start_index=ls
    res = np.zeros((ls*layers, ls*layers), np.int)
    for arr in layer_distance:
        res[cities_west_start_index: (cities_west_start_index+ls),
           cities_east_start_index: (cities_east_start_index+ls)] = arr
        cities_west_start_index += ls
        cities_east_start_index += ls

    # print(res)
    return res
```
