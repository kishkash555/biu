```python
seat_to_track = {1: 48, 2: 45, 3: 43}
df1
```

```python
mapping = {
        (int(gr_id), seat_to_track.get(int(seat))): par_id for seat, par_id,gr_id in df1.fillna(0).values
    }


del mapping[(1033,None)]
mapping[(1033, 45)] = 198

del mapping[(1021,None)]
mapping[(1021,43)] =163

mapping[(1028,48)] = 182
mapping[(1029,48)] = 185
mapping
```

```python
with open('../../../data/group_id_and_track_to_par_id.json','wt') as a:
    json.dump({ f"{gr}-{tr}":v for (gr,tr),v in mapping.items()},a)
```
