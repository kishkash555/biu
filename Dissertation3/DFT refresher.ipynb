```python
from numpy import pi
n=100
w = 5
y2 = np.cos(np.arange(n)*w*2*pi/n+0.1)#*np.hanning(n)
plt.plot(y2)
```

```python
f1 = np.fft.fft(y2)/n
fr = np.abs(f1)
fs = np.concatenate([fr[:1],fr[1:int(n/2)]+fr[n-1:int(n/2):-1]])
plt.plot(fs)
```

```python
print(np.array2string(np.vstack([np.arange(50),np.round(fs,2)]),suppress_small=True))
```
