```python
with open('../../../data/group_id_and_track_to_par_id.json','rt') as a:
    mapping = json.load(a)
mapping = {tuple(int(k) for k in key.split('-')) : v  for key, v in mapping.items() if not key.startswith('0')}
mapping
```

```python
sess_code = {'I': 'Interaction', 'FS': 'Freestyle' }
i = 0
for group_obj in svdata._data:
    group_id = group_obj._data['group_id']
    print(group_id)
    for sess in group_obj._data['sessions']._data:
        print(sess._data['session'])
        
        if sess._data['session'] in sess_code:
            sess_name = sess_code[sess._data['session']]
            try:
                drumming_data = drumming_df.set_index(['group','session']).loc[(group_id,sess_name),'drummer':'signal']
                sess._data['drumming'] = sv.list_store([
                    sv.dict_store({'drummer': r.drummer, 'participant_id': mapping[(group_id,r.drummer)] , 'signal_4hz': sv.list_shorthand(r.signal)})
                    for r in drumming_data.itertuples()
            ])
            except:
                continue
            #print(drumming_data)
            print(sess._data['drumming'])
        
    # if i==2: break
    i += 1
```
