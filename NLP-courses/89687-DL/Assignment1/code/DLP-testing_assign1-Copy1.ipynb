```python
# with a 'good' initialization, finds a solution with hidden layer of 2
params = mlp1.create_classifier(2,2,2)
params[0] = np.array([[1, -2],[-2,1]],np.double)
params[2] = np.array([[1,0],[0,1]],np.double)
tp1.train_classifier(xor_data1,xor_data1, 50, 0.1, params)
```

```python
# again random initialization, 8 nodes in hidden layer - this time took 28 iterations
params = mlp1.create_classifier(2,8,2)
params= [tp1.randomize_array(p) for p in params]
tp1.train_classifier(xor_data1,xor_data1, 50, 0.03, params)
```

```python
# random initialization, 2 nodes in hidden layer - not there even after 500 iterations
params = mlp1.create_classifier(2,2,2)
params= [tp1.randomize_array(p) for p in params]
tp1.train_classifier(xor_data1,xor_data1, 500, 0.03, params)
```

```python
# with the right initialization, find a solution with 2 nodes in hidden layer 
params = mlp1.create_classifier(2,2,2)
params[0] = np.array([[-2, 0],[8,-2]],np.double)
params[2] = np.array([[1, 0],[0,1]],np.double)

params= [tp1.randomize_array(p) for p in params]
tp1.train_classifier(xor_data1,xor_data1, 60, 0.08, params)
```
